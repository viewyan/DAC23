```python
import os
import pandas as pd
from collections import defaultdict
from pymatgen.core import Structure
from mendeleev import element

# 设置路径
folder_path = "/Users/viewyan/Documents/Datalab2025/pristine_CO2"

# 初始化缓存 + 属性收集容器
element_cache = {}
atom_data = defaultdict(list)

# 文件计数
counter = 0
max_files = 1000  # 先测试前1000个文件

# 开始循环
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith(".cif"):
            counter += 1
            if counter > max_files:
                break
            cif_path = os.path.join(root, file)
            try:
                structure = Structure.from_file(cif_path)
                for site in structure:
                    symbol = str(site.specie)

                    # 使用缓存
                    if symbol not in element_cache:
                        try:
                            element_cache[symbol] = element(symbol)
                        except Exception as e:
                            print(f"⚠️ Skipped element {symbol}: {e}")
                            element_cache[symbol] = None

                    el = element_cache[symbol]
                    if el is None:
                        continue

                    # 收集属性
                    atom_data["MOF"].append(file)
                    atom_data["Atomic symbol"].append(symbol)
                    atom_data["Atomic number"].append(el.atomic_number)
                    atom_data["Period"].append(el.period)
                    atom_data["Group"].append(el.group_id)
                    atom_data["Electronegativity"].append(el.en_pauling)
                    atom_data["Electron affinity"].append(el.electron_affinity)
                    atom_data["Melting point"].append(el.melting_point)
                    atom_data["Boiling point"].append(el.boiling_point)
                    ionization = el.ionenergies
                    atom_data["First ionization energy"].append(ionization.get(1, None) if ionization else None)
            except Exception as e:
                print(f"⛔ Error reading {file}: {e}")
                continue

df_atom_properties = pd.DataFrame(atom_data)

# df_atom_properties.to_csv("atom_properties_test.csv", index=False)

# 显示前几行
pd.set_option('display.max_columns', None)
print(df_atom_properties.columns.tolist())
pd.set_option('display.max_columns', None)  # 保证所有列都能显示
print("------ 前5行 Head ------")
print(df_atom_properties.head())

print("\n------ 后5行 Tail ------")
print(df_atom_properties.tail())

```

```python
import numpy as np

# 需要计算的字段
fields_to_process = [
    'Atomic number',
    'Period',
    'Group',
    'Electronegativity',
    'Electron affinity',
    'Melting point',
    'Boiling point',
    'First ionization energy'
]

# 正确写法
agg_dict = {}
for field in fields_to_process:
    agg_dict[field] = [
        ('mean', np.mean),
        ('geometric_mean', lambda x: np.exp(np.mean(np.log(x[x > 0])))),
        ('std', np.std),
        ('max', np.max),
        ('min', np.min)
    ]

# groupby 计算
grouped = df_atom_properties.groupby('MOF').agg(agg_dict)

# 展平 MultiIndex 列名
grouped.columns = [
    f"{col[0].lower().replace(' ', '_')}_{col[1]}"
    for col in grouped.columns
]

# 重置 index
grouped = grouped.reset_index()

# 看结果
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 2000)
pd.set_option('display.max_colwidth', None)

# 展示前 5 个 MOF
display(grouped.head(5))

# 保存成 CSV 文件
output_path = "/Users/viewyan/Documents/Datalab2025/pristine_CO2_fingerprints.csv"

grouped.to_csv(output_path, index=False)

print(f"Save to：{output_path}")


```

```python
import os
import random
from pymatgen.io.cif import CifParser

# Top-level folder path
folder_path = "/Users/viewyan/Documents/Datalab2025/pristine_CO2"

# List all subfolders
subfolders = [os.path.join(folder_path, d) for d in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, d))]

# Randomly pick 5 subfolders
sample_subfolders = random.sample(subfolders, 5)

print(f"\nSelected {len(sample_subfolders)} subfolders:")

for subdir in sample_subfolders:
    print(f"- {subdir}")

# Process each selected subfolder
for subdir in sample_subfolders:
    cif_files = [f for f in os.listdir(subdir) if f.endswith('.cif')]
    if cif_files:
        # Pick the first CIF file in this subfolder
        example_cif_path = os.path.join(subdir, cif_files[0])
        print(f"\n========================")
        print(f"Processing file: {example_cif_path}")

        try:
            parser = CifParser(example_cif_path)
            structure = parser.get_structures()[0]

            # Lattice parameters
            print("----- Lattice parameters -----")
            print(structure.lattice)

            # Space group
            print("\n----- Space group -----")
            try:
                print(parser.spacegroup)
            except:
                print("No spacegroup info")

            # Atomic species
            print("\n----- Atomic species -----")
            print([str(site.specie) for site in structure.sites])

            # Atomic positions (fractional)
            print("\n----- Atomic positions (fractional) -----")
            for site in structure.sites:
                print(f"{site.specie}: {site.frac_coords}")

            # Atomic positions (cartesian)
            print("\n----- Atomic positions (cartesian) -----")
            for site in structure.sites:
                print(f"{site.specie}: {site.coords}")

            # Occupancy
            print("\n----- Occupancy (if available) -----")
            for site in structure.sites:
                print(f"{site.specie}: occupancy = {site.properties.get('occupancy', 'N/A')}")

        except Exception as e:
            print(f"Error parsing file: {e}")
```
